**Loading dataset**

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/kaggle/input/crop-production-data-india/Crop Production data.csv')
df
```

**Asking and answer the question about the dat by visualization**

**1.Which states have the highest crop production?**

```python
state_production = df.groupby('State_Name')['Production'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(14, 6))
sns.barplot(data=state_production.head(10), x='State_Name', y='Production', palette='viridis')
plt.title('Top 10 States by Crop Production')
plt.xlabel('State')
plt.ylabel('Total Production')
plt.xticks(rotation=45)
plt.show()
```

**2.What are the top 10 crops in terms of total production?**

```python
crop_production = df.groupby('Crop')['Production'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(14, 6))
sns.barplot(data=crop_production.head(10), x='Crop', y='Production', palette='Blues_r')
plt.title('Top 10 Crops by Production')
plt.xlabel('Crop')
plt.ylabel('Total Production')
plt.xticks(rotation=45)
plt.show()
```

**3.Which District's  have the highest crop production?**


```python
district_production = df.groupby('District_Name')['Production'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(14, 6))
sns.barplot(data=district_production.head(10), x='District_Name', y='Production', palette='autumn')
plt.title('Top 10 Districts by Crop Production')
plt.xlabel('District')
plt.ylabel('Total Production')
plt.xticks(rotation=45)
plt.show()
```

**4.what is the Overall Distribution of the Top 10 crops that are produced all over India ?**

```python
crop_counts = df['Crop'].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.bar(crop_counts.index, crop_counts.values, color='gold', edgecolor='black')

plt.title('Overall Distribution of the Top 10 crops that are produced all over India', fontsize=12)

plt.xticks(crop_counts.index, crop_counts.index, rotation=90)

plt.xlabel('Name of the Crop')
plt.ylabel('Total Number of Values in the Data Frame')

plt.show()
```

**5. What is the total crop prodution based on the year?**

```python
plt.figure(figsize=(12, 6))
production_by_year = df.groupby('Crop_Year')['Production'].sum().reset_index()
sns.lineplot(data=production_by_year, x='Crop_Year', y='Production')
plt.title('Total Crop Production by Year')
plt.xlabel('Year')
plt.ylabel('Total Production')
plt.show()
```

**6. How is the crop prodccution during seasons?**

```python
groupby_crop = df.groupby("Season")["Production"].median().reset_index()
plt.figure(figsize=(15, 5))
sns.barplot(x=groupby_crop["Season"], y=groupby_crop["Production"], palette='Blues')
plt.title("Crops produced during different Seasons.")
plt.xlabel("Season")
plt.ylabel("Production")
plt.xticks(rotation=45) 
plt.show()
```

**Compare the Prodution by different states?**

```python
groupby_crop = df.groupby("State_Name")["Production"].median().reset_index().head(25)

plt.figure(figsize=(20,5))
sns.barplot(x=groupby_crop["State_Name"], y=groupby_crop["Production"], palette='plasma')
plt.title("compare the production across different States", fontsize=20)
plt.xlabel("State",  fontsize=12)
plt.ylabel("Production",  fontsize=12)
plt.xticks(rotation=45) 
plt.show()
```

**Conclusion**

The crop production analysis reveals several significant insights. Overall, crop production is heavily influenced by both seasonal and regional factors. Certain states consistently dominate production due to favorable climatic and geographical conditions. The analysis shows that a few key crops contribute disproportionately to the total production, highlighting their economic importance. Seasonal variations significantly impact production volumes, with some seasons showing markedly higher outputs. The correlation between cultivated area and production suggests that while larger areas generally lead to higher production, efficiency and yield per unit area are critical factors. Moreover, the distribution of production across different districts and the contribution of specific crops indicate a need for targeted agricultural policies to optimize resource allocation and enhance productivity in underperforming regions.
