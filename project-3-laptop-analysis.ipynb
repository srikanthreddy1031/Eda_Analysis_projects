```python

avg_price = df['Price_euros'].mean()
median_price = df['Price_euros'].median()
print(f'Average Price: {avg_price}, Median Price: {median_price}')
```

```python
# Correlation between weight and price
correlation = df['Weight'].corr(df['Price_euros'])
print(f'Correlation between Weight and Price: {correlation}')
```

```python
# brand and calculate average price by brand
avg_price_by_brand = df.groupby('Company')['Price_euros'].mean().sort_values(ascending=False)
print(avg_price_by_brand.head(10))  # Top 10 brands by average price
```

**Asking Questions based on Data and Visualization**

**1. Distrubiton of Laptops by RAM size**

```python
ram_distribution = df['Ram'].value_counts()
ram_distribution.plot(kind='bar', figsize=(10, 6),color='green')
plt.title('Distribution of Laptops by RAM Size')
plt.xlabel('RAM Size (GB)')
plt.ylabel('Number of Laptops')
plt.xticks(rotation=0)
plt.show()
```

**2.What is the distribution of different laptop types?**

```python
plt.figure(figsize=(10, 6))
type_counts = df['TypeName'].value_counts()
type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Laptop Types')
plt.ylabel('')  # Remove the y-label for a cleaner look
plt.show()
```

**3. Which GPU companies dominate the market in your dataset?**

```python
plt.figure(figsize=(12, 6))
gpu_counts = df['GPU_company'].value_counts()
gpu_counts.plot(kind='bar', color='orange')
plt.title('Count of Laptops by GPU Company')
plt.xlabel('GPU Company')
plt.ylabel('Number of Laptops')
plt.xticks(rotation=45)
plt.show()
```

**4.What isthe price of the  laptops based on Os by CPU_frequency?**

```python
plt.figure(figsize = (10,8))
sns.boxplot(x = df['OS'], y= df['Price_euros'])
plt.title('OS VS CPU_freqency')
plt.show()
```

**5.What isthe price of the  laptops based on Os by TouchScreen?**

```python
plt.figure(figsize = (10,8))
sns.barplot(x = df['OS'], y= df['Price_euros'] , hue = df['Touchscreen'])
plt.show()
```

**6. what is the Diatribution based on Ram size ?**

```python
ram_distribution = df['Ram'].value_counts()
ram_distribution.plot(kind='bar', figsize=(10, 6))
plt.title('Distribution of Laptops by RAM Size')
plt.xlabel('RAM Size (GB)')
plt.ylabel('Number of Laptops')
plt.xticks(rotation=0)
plt.show()
```

**CONCLUSION**

The analysis of the laptop dataset reveals significant insights into the market dynamics, highlighting the relationship between various specifications and pricing. Key findings include a notable correlation between RAM size and price, suggesting that higher RAM configurations tend to command higher prices. Additionally, the data indicates that certain brands, particularly those associated with premium components like high-end CPUs and GPUs, consistently exhibit higher average prices. The distribution of touchscreen laptops versus non-touchscreen options shows a growing preference for touchscreen technology, reflecting current consumer trends. Overall, the analysis underscores the importance of both technical specifications and brand reputation in influencing laptop pricing, providing valuable insights for both consumers and manufacturers in navigating the competitive landscape.
