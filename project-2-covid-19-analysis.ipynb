1. **Data Loading**

```python
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt


df_country=pd.read_csv('/kaggle/input/corona-virus-report/country_wise_latest.csv')
df_clean=pd.read_csv('/kaggle/input/corona-virus-report/covid_19_clean_complete.csv')
df_day=pd.read_csv('/kaggle/input/corona-virus-report/day_wise.csv')
df_usa=pd.read_csv('/kaggle/input/corona-virus-report/usa_county_wise.csv')
df_world=pd.read_csv('/kaggle/input/corona-virus-report/worldometer_data.csv')
df_group=pd.read_csv('/kaggle/input/corona-virus-report/full_grouped.csv')
```

**Merging Datasets**

```python
covid_columns = set(df_clean.columns)
country_columns = set(df_country.columns)
world_column=set(df_world.columns)
common_columns = covid_columns & country_columns
print(common_columns)
```

```python
commons = ['WHO Region', 'Recovered', 'Country/Region', 'Deaths', 'Active', 'Confirmed']
merge = pd.merge(df_clean, df_country, on=commons, how='inner')
merge.head()
```

**Asking Question About datasets**

**1.What is the distribution of total deaths across different WHO Regions?**

```python

region_deaths = merge.groupby('WHO Region')['Deaths'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='WHO Region', y='Deaths', data=region_deaths)
plt.xlabel('WHO Region')
plt.ylabel('Total Deaths')
plt.title('Distribution of Total Deaths Across WHO Regions')
plt.xticks(rotation=45)
plt.show()
```

****2.Which countries have seen the highest increase in confirmed cases in the last week?****

```python
highest_increase = merge[['Country/Region', '1 week % increase']].drop_duplicates().sort_values(by='1 week % increase', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='1 week % increase', y='Country/Region', data=highest_increase, palette='viridis')
plt.xlabel('1 Week % Increase')
plt.ylabel('Country/Region')
plt.title('Top 10 Countries with Highest Increase in Confirmed Cases in the Last Week')
plt.show()
```

**3.What is the recovery rate in different countries?**

```python
merge['Recovery Rate'] = merge['Recovered'] / merge['Confirmed'] * 100

latest_data = merge[merge['Date'] == merge['Date'].max()]

top_recovery_countries = latest_data[['Country/Region', 'Recovery Rate']].sort_values(by='Recovery Rate', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='Recovery Rate', y='Country/Region', data=top_recovery_countries, palette='Blues_d')
plt.xlabel('Recovery Rate (%)')
plt.ylabel('Country/Region')
plt.title('Top 10 Countries by Recovery Rate')
plt.show()
```

**4.What is the proportion of deaths in the top 5 countries by total deaths?**

```python
top_death_countries = merge.groupby('Country/Region')['Deaths'].sum().sort_values(ascending=False).head(10).reset_index()

plt.figure(figsize=(8, 8))
plt.pie(top_death_countries['Deaths'], labels=top_death_countries['Country/Region'], autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Deaths in Top 5 Countries by Total Deaths')
plt.axis('equal')
plt.show()
```

**5.Global Distribution of COVID-19 Cases (World Map)**

```python
import plotly.express as px

global_data = merge.groupby('Country/Region')[['Confirmed', 'Deaths', 'Recovered']].sum().reset_index()

fig = px.scatter_geo(global_data, locations="Country/Region", locationmode='country names', 
                     size="Confirmed", color="Deaths", hover_name="Country/Region",
                     title='Global Distribution of COVID-19 Cases',
                     projection="natural earth")
fig.show()
```

**6. Total Confirmed Cases by Continent**

```python

continent_confirmed = df_world.groupby('Continent')['TotalCases'].sum().reset_index()
colors = ['skyblue', 'orange', 'green', 'red', 'purple', 'brown', 'pink']

plt.figure(figsize=(10, 6))
plt.bar(continent_confirmed['Continent'], continent_confirmed['TotalCases'], color=colors)
plt.xlabel('Continent')
plt.ylabel('Confirmed Cases')
plt.title('Total Confirmed Cases by Continent')
plt.show()
```
