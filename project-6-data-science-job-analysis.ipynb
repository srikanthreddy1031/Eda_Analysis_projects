**Data Loading**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('/kaggle/input/data-science-jobs-salaries/Data Science Jobs Salaries.csv')
df
```

```python
df['work_year'] = df['work_year'].str.extract('(\d+)')
df['work_year']=df['work_year'].astype(int)
```

**1.What is the Avg Salry in USD based on years?**

```python
avg_salary_per_year = df.groupby('work_year')['salary_in_usd'].mean()
plt.figure(figsize=(10, 6))
plt.plot(avg_salary_per_year.index, avg_salary_per_year.values, marker='o', linestyle='-', color='b')
plt.title('Average Salary in USD Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Salary in USD')
plt.grid(True)
plt.show()
```

**2.What is the distribution of job titles among data science professionals?**

```python
job_title_counts = df['job_title'].value_counts()

plt.figure(figsize=(12, 6))
job_title_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Job Titles Among Data Science Professionals')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**3. what is salary Distribution based on the experience level?**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df)
plt.title('Salary Distribution by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary in USD')
plt.show()
```

**4. Which job titles offer the highest average salary?**

```python
avg_salary_per_job_title = df.groupby('job_title')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
avg_salary_per_job_title.plot(kind='bar', color='salmon')
plt.title('Top 10 Job Titles with Highest Average Salary')
plt.xlabel('Job Title')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**5.How does the average salary in USD vary across different countries of employee residence?**

```python
avg_salary_per_residence = df.groupby('employee_residence')['salary_in_usd'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
avg_salary_per_residence.plot(kind='bar', color='lightblue')
plt.title('Top 10 Countries by Average Salary in USD')
plt.xlabel('Employee Residence')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**6.What is the distribution of employment types across different job titles?**

```python
employment_job_title = df.pivot_table(index='job_title', columns='employment_type', aggfunc='size', fill_value=0)
employment_job_title.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='Paired')
plt.title('Distribution of Employment Types Across Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Employment Type')
plt.show()
```

**Conclusion**

The analysis of data science jobs and salaries reveals several key trends. Average salaries in USD have shown an upward trend over the years, reflecting the increasing demand and value of data science roles. There is a significant variation in salaries based on experience level, with senior professionals earning substantially more than entry-level ones. Certain job titles, such as 'Data Engineer' and 'Machine Learning Engineer', command higher salaries, indicating their specialized skills and demand. The employment type distribution shows a prevalence of full-time roles, while remote work ratios demonstrate that companies with higher remote work flexibility tend to offer competitive salaries. Geographically, employees in countries like the United States and Switzerland receive higher compensation. Additionally, larger companies offer more consistent salaries, suggesting greater resources and standardized pay scales. Overall, the data highlights the importance of experience, job specialization, and location in determining salaries within the data science field.
